```python
from requests import get, post
import urllib
import urllib.parse
from bs4 import BeautifulSoup
from IPython.core.display import display, HTML
import re
from time import sleep
```

```python
def remove_all_brackets(s):
    stack = []
    result = ''
    openers = ['(', '[']
    closers = [')', ']']
    first = 0
    for i in range(len(s)):
        if s[i] in openers:
            if not len(stack):
                result += s[first : i]
            stack.append(s[i])
        elif s[i] in closers:
            stack = stack[:-1]
            if not len(stack):
                first = i + 1
    if not len(stack):
        result += s[first :]
    return result
```

```python
def reach_philosophy(url, depth = 10):
    titles_list = []
    new_url = url
    
    while True:
#         sleep(0.5)
        req = get(new_url)
        soup = BeautifulSoup(req.text)
        
        title = soup.findAll('h1', attrs={'id': 'firstHeading'})[0].text
            
        if title in titles_list:
#             print("зациклились на странице {}".format(title))
            return (False, "зациклились на странице '{}'".format(title), titles_list + [title])
        
        titles_list.append(title)
        if title in ['Philosophy', 'Философия']:
            return (True, titles_list)

        try:
            div = soup.findAll('div', recursive=True, attrs={'class': 'mw-parser-output'})[0]
        except IndexError:
            return (False, 0, [])
        
        new_url = None
        for p in div.findAll(['p', 'ul', 'ol'], recursive=False):
            par = BeautifulSoup(remove_all_brackets(str(p)))
            refs = par.findAll('a')
            for a in refs:
                href = a.get('href')
                if href and a.get('title') and (a.get('class') != 'new'):
                    new_url = 'https://ru.wikipedia.org' + href
                    break
            if new_url:
                break
        if not new_url:
            return (False, "нет ссылок на странице '{}'".format(title), titles_list)
```

```python
reach_philosophy('https://ru.wikipedia.org/wiki/%D0%90%D1%88%D0%BE%D1%82_I')
```
